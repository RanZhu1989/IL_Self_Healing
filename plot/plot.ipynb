```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bc_rate = np.loadtxt('bc.csv',delimiter=',')
bc_max = np.max(bc_rate,axis=1)
bc_min = np.min(bc_rate,axis=1)
bc_mean = np.mean(bc_rate,axis=1)
dqn_rate = np.loadtxt('dqn.csv',delimiter=',')
dqn_max = np.max(dqn_rate,axis=1)
dqn_min = np.min(dqn_rate,axis=1)
dqn_mean = np.mean(dqn_rate,axis=1)
ppo_rate = np.loadtxt('ppo.csv',delimiter=',')
ppo_max = np.max(ppo_rate,axis=1)
ppo_min = np.min(ppo_rate,axis=1)
ppo_mean = np.mean(ppo_rate,axis=1)
```

```python
plt.plot(bc_mean, label='BC',color='red',linewidth=2)
plt.fill_between(np.arange(0,200),bc_min,bc_max,color='red',alpha=0.2)
plt.plot(dqn_mean, label='DQN',color='blue',linewidth=2)
plt.fill_between(np.arange(0,200),dqn_min,dqn_max,color='blue',alpha=0.2)
plt.plot(ppo_mean, label='PPO',color='green',linewidth=2)
plt.fill_between(np.arange(0,200),ppo_min,ppo_max,color='green',alpha=0.2)
plt.xlabel('Episode')
plt.ylabel('Success Ratio')
plt.legend()
plt.show()
```

```python
import csv
import numpy as np

data = []

# CSV
with open('data2.csv', 'r') as file:
    # CSV
    csv_reader = csv.reader(file, delimiter=',')

    # 
    for row in csv_reader:
        row_data = []
        for element in row:
            # 
            element_values = element.strip("[]").split()
            # 
            element_data = [float(value) for value in element_values]
            row_data.append(element_data)
        data.append(row_data)

# NumPy
data = np.array(data)

# 
print(data)
```
