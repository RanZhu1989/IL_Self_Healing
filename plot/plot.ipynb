```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
N-1
'''
tieline_bc1_rate = np.loadtxt('./n_1/TieLine_BC.csv',delimiter=',')
tieline_bc1_max = np.max(tieline_bc1_rate,axis=1)
tieline_bc1_min = np.min(tieline_bc1_rate,axis=1)
tieline_bc1_mean = np.mean(tieline_bc1_rate,axis=1)

hybrid_bc1_rate = np.loadtxt('./n_1/Hybrid_BC.csv',delimiter=',')
hybrid_bc1_max = np.max(hybrid_bc1_rate,axis=1)
hybrid_bc1_min = np.min(hybrid_bc1_rate,axis=1)
hybrid_bc1_mean = np.mean(hybrid_bc1_rate,axis=1)

tieline_dagger1_rate = np.loadtxt('./n_1/TieLine_DAgger.csv',delimiter=',')
tieline_dagger1_max = np.max(tieline_dagger1_rate,axis=1)
tieline_dagger1_min = np.min(tieline_dagger1_rate,axis=1)
tieline_dagger1_mean = np.mean(tieline_dagger1_rate,axis=1)

hybrid_dagger1_rate = np.loadtxt('./n_1/Hybrid_DAgger.csv',delimiter=',')
hybrid_dagger1_max = np.max(hybrid_dagger1_rate,axis=1)
hybrid_dagger1_min = np.min(hybrid_dagger1_rate,axis=1)
hybrid_dagger1_mean = np.mean(hybrid_dagger1_rate,axis=1)

plt.figure(figsize=(10,5))

plt.plot(tieline_bc1_mean, label='Tieline_BC',color='red',linewidth=0.5)
plt.fill_between(np.arange(0,500),tieline_bc1_min,tieline_bc1_max,color='red',alpha=0.1)
plt.plot(hybrid_bc1_mean, label='Hybrid_BC',color='blue',linewidth=0.5)
plt.fill_between(np.arange(0,500),hybrid_bc1_min,hybrid_bc1_max,color='blue',alpha=0.1)
plt.plot(tieline_dagger1_mean, label='TieLine_DAgger',color='green',linewidth=1)
plt.fill_between(np.arange(0,500),tieline_dagger1_min,tieline_dagger1_max,color='green',alpha=0.1)
plt.plot(hybrid_dagger1_mean, label='Hybrid_Dagger',color='orange',linewidth=1)
plt.fill_between(np.arange(0,500),hybrid_dagger1_min,hybrid_dagger1_max,color='orange',alpha=0.1)


plt.title('N-1 Success Ratio')
plt.xlabel('Episode')
plt.ylabel('Success Ratio')
plt.legend()
plt.savefig('../pics/N-1_Success_Ratio.png', dpi=300, bbox_inches='tight')
plt.show()
```
