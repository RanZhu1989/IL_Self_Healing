```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def _moving_average(data, window_size):
    """
    Calculate the moving average of a 1D array.
    """
    pad_size = window_size // 2
    padded_data = np.pad(data, (pad_size, pad_size), mode='edge')
    window = np.ones(int(window_size)) / float(window_size)
    smoothed_data = np.convolve(padded_data, window, 'same')
    return smoothed_data[pad_size:-pad_size]
```

```python
'''
N-1
'''
tieline_bc1_rate = np.loadtxt('./n_1/TieLine_BC.csv',delimiter=',')
tieline_bc1_std = _moving_average(np.std(tieline_bc1_rate,axis=1),50)
tieline_bc1_mean = _moving_average(np.mean(tieline_bc1_rate,axis=1),50)

hybrid_bc1_rate = np.loadtxt('./n_1/Hybrid_BC.csv',delimiter=',')
hybrid_bc1_std = _moving_average(np.std(hybrid_bc1_rate,axis=1),50)
hybrid_bc1_mean = _moving_average(np.mean(hybrid_bc1_rate,axis=1),50)

tieline_dagger1_rate = np.loadtxt('./n_1/TieLine_DAgger.csv',delimiter=',')
tieline_dagger1_std = _moving_average(np.std(tieline_dagger1_rate,axis=1),50)
tieline_dagger1_mean = _moving_average(np.mean(tieline_dagger1_rate,axis=1),50)

hybrid_dagger1_rate = np.loadtxt('./n_1/Hybrid_DAgger.csv',delimiter=',')
hybrid_dagger1_std = _moving_average(np.std(hybrid_dagger1_rate,axis=1),50)
hybrid_dagger1_mean = _moving_average(np.mean(hybrid_dagger1_rate,axis=1),50)

tieline_gail_rate = np.loadtxt('./n_1/TieLine_GAIL.csv',delimiter=',')
tieline_gail_std = _moving_average(np.std(tieline_gail_rate,axis=1),100)
tieline_gail_mean = _moving_average(np.mean(tieline_gail_rate,axis=1),100)

plt.figure()
plt.plot(tieline_bc1_mean, label='Tieline_BC',linewidth=2)
plt.fill_between(np.arange(0,500),tieline_bc1_mean-tieline_bc1_std,tieline_bc1_mean+tieline_bc1_std,alpha=0.3)
plt.plot(tieline_dagger1_mean, label='TieLine_DAgger',linewidth=2)
plt.fill_between(np.arange(0,500),tieline_dagger1_mean-tieline_dagger1_std,tieline_dagger1_mean+tieline_dagger1_std,alpha=0.3)
plt.title('N-1 Success Ratio')
plt.xlabel('Epoch')
plt.ylabel('Success Ratio')
plt.legend()
plt.savefig('../pics/N-1_Success_Ratio_A.png', dpi=300, bbox_inches='tight')
plt.show()

plt.figure()
plt.plot(hybrid_bc1_mean, label='Hybrid_BC',linewidth=2)
plt.fill_between(np.arange(0,500),hybrid_bc1_mean-hybrid_bc1_std,hybrid_bc1_mean+hybrid_bc1_std,alpha=0.3)
plt.plot(hybrid_dagger1_mean, label='Hybrid_Dagger',linewidth=2)
plt.fill_between(np.arange(0,500),hybrid_dagger1_mean-hybrid_dagger1_std,hybrid_dagger1_mean+hybrid_dagger1_std,alpha=0.3)
plt.title('N-1 Success Ratio')
plt.xlabel('Epoch')
plt.ylabel('Success Ratio')
plt.legend()
plt.savefig('../pics/N-1_Success_Ratio_B.png', dpi=300, bbox_inches='tight')
plt.show()

plt.figure()
plt.plot(tieline_gail_mean, label='Tieline_GAIL',linewidth=2)

plt.fill_between(np.arange(0,5000),tieline_gail_mean-tieline_gail_std,tieline_gail_mean+tieline_gail_std,alpha=0.3)
plt.title('N-1 Success Ratio')
plt.xlabel('Epoch')
plt.ylabel('Success Ratio')
plt.legend()
```
