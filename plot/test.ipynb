```python
import csv
import numpy as np

data = []

# 打开CSV文件
with open('data2.csv', 'r') as file:
    # 创建CSV读取器
    csv_reader = csv.reader(file, delimiter=',')

    # 逐行读取数据
    for row in csv_reader:
        row_data = []
        for element in row:
            # 去除中括号，并按空格分割字符串
            element_values = element.strip("[]").split()
            # 将每个值转换为浮点数
            element_data = [float(value) for value in element_values]
            row_data.append(element_data)
        data.append(row_data)

# 将数据转换为NumPy数组
data = np.array(data)

# 打印结果
print(data)
```
